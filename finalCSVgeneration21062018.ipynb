```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

Header creation

```python
subjects = ['0025427', '0025428', '0025429', '0025430', '0025431', '0025433', '0025434', '0025435', '0025436', '0025437', 
            '0025438', '0025440', '0025441', '0025442', '0025444', '0025445', '0025446', '0025447', '0025448', '0025450', 
            '0025451', '0025452', '0025453', '0025454', '0025455', '0025456', 'Chum', 'Chus', 'Cinq', 'Ubc', 'Ismd', 'Iugm',
            'Mni', 'Edm', 'Rri', 'Sask', 'Sun', 'Twh', 'Vic']
headers = []
#HNU
for s in range(26):
    for k in range(0,10):
        headers.append('s' + subjects[s] + '-' + str(k))
#SSimon        
for s in range(26,39):
    #Session 1
    headers.append(subjects[s] + '-0')
    
    #Session2
    if ((s != 29) and (s < 33)):
        headers.append(subjects[s] + '-1')
    
    #Session3
    if (s != 29 and s != 26 and s < 33):
        headers.append(subjects[s] + '-2')
        
    #MNI rest2
    if (s == 32):
        headers.append(subjects[s] + '-3')
```

Loading days apart csv file for CIMAQ

```python
daysCIMAQ = pd.read_csv('CIMAQ-daysApart.csv', usecols=[1])['DaysApart'].tolist()
```

Lists used in final dataframe

```python
networks = ['DMN', 'MOT', 'SAL', 'VIS', 'LIM', 'FPN', 'CER']
names = []
db = pd.DataFrame(columns = ['network', 'pair', 'correlation', 'days_apart', 'site_label', 'site_label_concat', 
                             'exact_site_label1', 'exact_site_label2','scanner1', 'scanner2', 'scanner_Label', 
                             'visit1','subject1', 'sex1', 'visit2', 'subject2', 'sex2', 'id1', 'id2'],
                              index = range(0))
scanner = ['GE', 'Siemens', 'Philips']
scanLabel = 'GEGE'
```

Final dataframe

```python
for n in range(7):
    names.append("Corr/HnuSsimon-" + networks[n] + "-06182018.csv")

    data = pd.read_csv(names[n], header = None)
    data.columns = headers
    data.index = headers
    data.values[np.tril_indices_from(data, 0)] = np.nan
    data = data.round(3);
    df = pd.DataFrame(columns = ['network', 'pair', 'correlation', 'days_apart', 'site_label', 'site_label_concat', 
                             'exact_site_label1', 'exact_site_label2','scanner1', 'scanner2', 'scanner_Label', 
                             'visit1','subject1', 'sex1', 'visit2', 'subject2', 'sex2', 'id1', 'id2'],index = range(0,81225))

    a = 0 #counter final dataFrame
    c = 0 #counter CIMAQ vs CIMAQ daysApart
    for i in range(285):
        for j in range(285): 
            #Network
            df.values[a,0] = networks[n]                                               
            
            #Sites
            df.values[a,1] = data.index[i] + ' vs ' + data.columns[j]                  
            
            #Corr
            df.values[a,2] = data.values[i,j]                                          
            
            #Days Apart HNU vs HNU
            if (i <= 259 and j <= 259):
                df.values[a,3] = abs((int(data.columns[j][-1]) - int(data.index[i][-1]))*3)    
            
            #Days Apart CIMAQ vs CIMAQ
            elif (i > 259 and j > 259):
                if (math.isnan(df.values[a,2]) == False):
                    df.values[a,3] = daysCIMAQ[c]                                              
                    c = c + 1
                    
            #Days Apart CIMAQ vs HNU
            else:
                df.values[a,3] = np.nan                                                        
            
            #Visit1
            df.values[a,11] = int(data.index[i][-1]) + 1                                
            
            #Subject1
            if (data.index[i][0] == 's'):
                sub1 = df.values[a,12] = data.index[i].partition("-")[0]     
            else:
                sub1 = df.values[a,12] = 'Csub'
                
            #sex1
            if (sub1 == 'Csub' or sub1 == 's0025427' or sub1 == 's0025433' or sub1 == 's0025434' or sub1 == 's0025436'
               or sub1 == 's0025437' or sub1 == 's0025440' or sub1 == 's0025446' or sub1 == 's0025447' or sub1 == 's0025448'
               or sub1 == 's0025451' or sub1 == 's0025454' or sub1 == 's0025455'):
                df.values[a,13] = 'M'
            else:
                df.values[a,13] = 'F'
                
            #Visit2
            df.values[a,14] = int(data.columns[j][-1]) + 1                              
            
            #Subject2
            if (data.columns[j][0] == 's'):
                sub2 = df.values[a,15] = data.columns[j].partition("-")[0]    
            else:
                sub2 = df.values[a,15] = 'Csub'
            
            #sex2
            if (sub2 == 'Csub' or sub2 == 's0025427' or sub2 == 's0025433' or sub2 == 's0025434' or sub2 == 's0025436'
               or sub2 == 's0025437' or sub2 == 's0025440' or sub2 == 's0025446' or sub2 == 's0025447' or sub2 == 's0025448'
               or sub2 == 's0025451' or sub2 == 's0025454' or sub2 == 's0025455'):
                df.values[a,16] = 'M'
            else:
                df.values[a,16] = 'F'
          
            #id1
            id1 = df.values[a,17] = data.index[i]  
            id1 = id1.partition("-")[0]
            
            #id2
            id2 = df.values[a,18] = data.columns[j]                                          
            id2 = id2.partition("-")[0]
            
            #site_label
            if (sub1[0] == 's'):
                siteLabel1 = df.values[a,4] = 'Hnu'
            else:
                siteLabel1 = df.values[a,4] = data.index[i].partition("-")[0]
            
            #site_label_concat
            if (sub2[0] == 's'):
                siteLabel2 = 'Hnu'
            else:
                siteLabel2 = data.columns[j].partition("-")[0]   
            df.values[a,5] = siteLabel1 + siteLabel2
        
            #exact_site_label1
            df.values[a,6] = siteLabel1 + data.index[i].partition("-")[2][:-1]  
            
            #exact_site_label2
            df.values[a,7] = siteLabel2 + data.columns[j].partition("-")[2][:-1]
             
            #Scanner1
            if (sub1[0] == 's' or id1 == 'Twh' or id1 == 'Vic'):
                df.values[a,8] = 'GE'                                                 
            elif (id1 == 'Chum' or id1 == 'Chus' or id1 == 'Cinq' or id1 == 'Ubc'):
                df.values[a,8] = 'Philips'                                             
            else:
                df.values[a,8] = 'Siemens'
                  
            #Scanner2
            if (sub2[0] == 's' or id2 == 'Twh' or id2 == 'Vic'):
                df.values[a,9] = 'GE'                                                  
            elif (id2 == 'Chum' or id2 == 'Chus' or id2 == 'Cinq' or id2 == 'Ubc'):
                df.values[a,9] = 'Philips'                                             
            else:
                df.values[a,9] = 'Siemens'
                  
            #Scanner Label
            df.values[a,10] = df.values[a,8] + df.values[a,9]                           
            
            a = a + 1
               
               
    db = pd.concat([db, df])
    
```

```python
dbx = db[db.correlation.notnull()] ##Drop lower triangle and diagonal

dbNoMni2 = dbx[dbx.id1 != 'Mni-2']
dbNoMni2 = dbNoMni2[dbNoMni2.id2 != 'Mni-2']

dbNoMni2.to_csv('HnuCsub-062262018.csv', index = False)
```
