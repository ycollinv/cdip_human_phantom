## 1A. Overall consistency (CSub)

#### i.Fetch CDIP consistency data from dataframe (df) to a new dataframe (df_Csub)

```python
df_Csub=df.loc[(df['subject1'] == 'Csub') & (df['subject1'] == df['subject2'])]
df_Csub
#Should have 23*24*7/2 = 1932 correlations
```

#### ii. Calculate mean (std) overall Csub consistency for all 7 networks


```python
df_Csub.groupby(["network"]).agg({"correlation": ["mean", "std"]}).reset_index()
```

#### iii. Calculate mean (std) overall Csub consistency across all 7 networks

```python
df_Csub.agg({"correlation": ["mean", "std"]}).reset_index()
```
